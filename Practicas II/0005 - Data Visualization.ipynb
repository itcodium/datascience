#### 5. Creating multiple plots

```python
# https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.arange.html
# numpy.arange(start, stop, step, dtype)

p1 = np.arange(0.0001, 30.0, 0.1)

plt.subplot(211)
plt.plot(p1, np.sin(p1)/p1, 'b--')

plt.subplot(212)
plt.plot(p1, np.cos(p1), 'r--')

plt.show()
```

#### 6. Playing with text

```python
# Playing with text
n = np.random.random_sample((5,))
plt.bar(np.arange(len(n)), n)
plt.xlabel('Indices')
plt.ylabel('Value')
plt.text(1, .7, r'$\mu=' + str(np.round(np.mean(n), 2)) + ' $')
plt.show()
```

```python
ax = plt.subplot(111)
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)

line, = plt.plot(t, s, lw=2)
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5), arrowprops=dict(facecolor='black', shrink=0.05),)
plt.ylim(-2,2)
plt.show()
```

#### 8. Box plots

```python
## Creating some data
np.random.seed(10)
box_data_1 = np.random.normal(100, 10, 200)
box_data_2 = np.random.normal(80, 30, 200)
box_data_3 = np.random.normal(90, 20, 200)

## Combining the different data in a list
data_to_plot = [box_data_1, box_data_2, box_data_3]

# Create the boxplot
bp = plt.boxplot(data_to_plot)
```

```python
## add patch_artist=True option to ax.boxplot()
## to get fill color
bp = plt.boxplot(data_to_plot, patch_artist=True)

for box in bp['boxes']:
    box.set( color='#7570b3', linewidth=2)
    box.set( facecolor = '#1b9e77' )

for whisker in bp['whiskers']:
    whisker.set(color='#7570b3', linewidth=2)

for cap in bp['caps']:
    cap.set(color='#7570b3', linewidth=2)

for median in bp['medians']:
    median.set(color='#b2df8a', linewidth=2)

for flier in bp['fliers']:
    flier.set(marker='o', color='#e7298a', alpha=0.5)
```
