# Seaborn

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
x = np.linspace(0, 5, 50)

y_cos = np.cos(x)
y_sin = np.sin(x)
plt.figure()
plt.plot(x,y_cos)
plt.plot(x,y_sin)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin/cos functions')
plt.show()
```

```python
x = np.linspace(0, 5, 50)
y_cos = np.cos(x)
y_sin = np.sin(x)

sns.set_context("talk")
sns.set_style('whitegrid')
with sns.axes_style('whitegrid'):
    plt.figure()
    
plt.plot(x,y_cos)
plt.plot(x,y_sin)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin/cos functions')
plt.show()
```
