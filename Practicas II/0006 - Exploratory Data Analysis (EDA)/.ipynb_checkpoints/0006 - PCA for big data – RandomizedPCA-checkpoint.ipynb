# Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt

iris = datasets.load_iris()

pca_2c = PCA(n_components=2)
X_pca_2c = pca_2c.fit_transform(iris.data)
X_pca_2c.shape
```

```python
plt.scatter(X_pca_2c[:,0], X_pca_2c[:,1], c=iris.target,alpha=0.8, s=60, marker='o', edgecolors='white')
plt.show()
```

```python
pca_2c.explained_variance_ratio_.sum()
```

```python
pca_2c.components_
```

### whiten the signal (or make it more spherical)

- Sometimes, you will find yourself in a situation where PCA is not effective enough, especially when dealing with high dimensionality data, since the features may be very correlated and, at the same time, the variance is unbalanced

```python
pca_2cw = PCA(n_components=2, whiten=True)
X_pca_1cw = pca_2cw.fit_transform(iris.data)
plt.scatter(X_pca_1cw[:,0], X_pca_1cw[:,1], c=iris.target,
alpha=0.8, s=60, marker='o', edgecolors='white')
plt.show()
```

```python
pca_2cw.explained_variance_ratio_.sum()
```

### Project the input dataset on an 1D space generated

```python
pca_1c = PCA(n_components=1)
X_pca_1c = pca_1c.fit_transform(iris.data)
plt.scatter(X_pca_1c[:,0], np.zeros(X_pca_1c.shape), c=iris.target,
alpha=0.8, s=60, marker='o', edgecolors='white')
plt.show()
```

### Generate an output set containing at least 95 percent of the input energy

- To ensure that you generate an output set containing at least 95 percent of the input energy, you can just specify this value to the PCA object during its first call

```python
pca_95pc = PCA(n_components=0.95)
X_pca_95pc = pca_95pc.fit_transform(iris.data)
print (pca_95pc.explained_variance_ratio_.sum())
print (X_pca_95pc.shape)
```
