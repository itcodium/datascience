# Normal distribution

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
x = [0, 1, 2, 3, 4, 5, 6]

# n: cantidad de puntos
# p: probability of success

n, p = 6, 0.5

# binom function takes in the number of attempts and the probability of success
rv = binom(n, p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,label='Probablity')
ax.legend(loc='best', frameon=False)
plt.show()
```

### 2. Normal distribution, increase the number of attempts

```python
fig, ax = plt.subplots(1, 1)
x = range(1000)

# n: cantidad de puntos
# p: probability of success

n, p = 1000, 0.5
rv = binom(n, p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,label='Probablity')
ax.legend(loc='best', frameon=False)
plt.show()
```
