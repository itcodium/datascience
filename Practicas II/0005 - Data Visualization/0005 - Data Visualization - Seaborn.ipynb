#### Enhancing your EDA capabilities

EDA to help you discover distributions and relationships between variables.

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.datasets import load_boston

iris = load_iris()
boston = load_boston()

X_iris, y_iris = iris.data, iris.target
X_boston, y_boston = boston.data, boston.target
```

```python
 {j: flower for j, flower in enumerate(iris.target_names)}
```

```python
features_iris = [a[:-5].replace(' ','_') for a in iris.feature_names]
target_labels = {j: flower for j, flower in enumerate(iris.target_names)}
df_iris = pd.DataFrame(X_iris, columns=features_iris)
df_iris['target'] = [target_labels[y] for y in y_iris]

df_iris.head()

```

```python
rBostonFN=range(len(boston.feature_names))
boston_mapped= zip(boston.feature_names,rBostonFN)
# set(boston_mapped) #Visualizar datos zippeados

features_boston = np.array(['V'+'_'.join([str(b), a]) for a,b in boston_mapped])

df_boston = pd.DataFrame(X_boston, columns=features_boston)
df_boston['target'] = y_boston
df_boston['target_level'] = pd.qcut(y_boston,3)

df_boston.head()
```

```python
with sns.axes_style("ticks"):
    sns.factorplot(data=df_boston, x='V8_RAD', y="target")
    
#g = sns.catplot(x="V8_RAD", y="target", kind="point", data=df_boston)    
 
```

#### scatterplot incorporating a regression fitted curve trend

```python
with sns.axes_style("whitegrid"):
    sns.regplot(data=df_boston, x='V12_LSTAT', y="target", order=3)
```

#### Where it is necessary to consider distributions too, jointplot will provide additional plots on the side of the scatterplot:

```python
# kind must be either 'scatter', 'reg', 'resid', 'kde', or 'hex'

with sns.axes_style("whitegrid"):
    sns.jointplot("V4_NOX", "V7_DIS", data=df_boston, kind='reg', order=3)
```

#### When the purpose is to discover what discriminates classes, FacetGrid can arrange different plots in a comparable way and help you understand where there are differences.

```python
with sns.axes_style("darkgrid"):
    chart = sns.FacetGrid(df_iris, col="target")
    chart.map(plt.scatter, "sepal_length", "petal_length")
```

```python
#with sns.axes_style("darkgrid"):
#    chart = sns.FacetGrid(df_iris, col="target")
#    chart.map(sns.distplot, "sepal_length")

g = sns.FacetGrid(df_iris, col="target")
g.map(sns.distplot, "sepal_length",  color="r");
```

```python
from scipy import stats
g = sns.FacetGrid(df_iris, col="target")
g.map(sns.distplot, "sepal_length",  color="r" , kde=False, rug=False, fit=stats.gamma);
```
