```python
import urllib3
import pandas as pd
 

modals_url="http://samplecsvs.s3.amazonaws.com/Sacramentorealestatetransactions.csv"
http = urllib3.PoolManager()
r = http.request('GET', modals_url)
response=r.data.decode("utf-8")

lines=response.split('\r')

lines

cols_name=lines[0].split(",")
n_cols=len(cols_name)
print("len ->",len(cols_name))
print("data -> ",cols_name)

counter=0
main_dict={}
for col in cols_name:
    main_dict[col]=[]
    

#reader_csv = csv.reader(response.split('\r'), delimiter=',')
for line in lines:
    if(len(line))>0:
        if (counter>0):
            values=line.strip().split(',')
            for i in range(len(cols_name)):
                main_dict[cols_name[i]].append(values[i])
        counter+=1
print("el dataset tiene %d filas y %d columnas" % (counter, n_cols))

data=pd.DataFrame(main_dict)
data.head()
```
