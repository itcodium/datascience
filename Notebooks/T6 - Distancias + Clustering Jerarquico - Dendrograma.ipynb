```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
path_root="C:/Users/MYEQ/Documents/Itcodium.git/datascience/Datasets/"
path=path_root+"movies/movies.csv"
data=pd.read_csv(path,sep=";")
data.head()
```

```python
movies=data.columns.values.tolist()[1:]
movies
```

```python
dd1=distance_matrix(data[movies],data[movies],p=1)
dd2=distance_matrix(data[movies],data[movies],p=2)
dd10=distance_matrix(data[movies],data[movies],p=10)
```

```python
def dm_to_df(dd,col_name):
    import pandas as pd
    return pd.DataFrame(dd,index=col_name,columns=col_name)
```

## Clustering Jerarquico - Dendrograma

```python
import matplotlib.pyplot  as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z=linkage(data[movies],"single") # "ward", "average", "complete"
Z
```

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerarquico para el clustering")
plt.xlabel("Id de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z,leaf_rotation=90.,leaf_font_size=10.) # ,orientation="right"
plt.show()
```
